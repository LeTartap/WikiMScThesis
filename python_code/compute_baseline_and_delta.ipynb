```python
import pandas as pd

tiny_revs =  pd.read_csv("after_spacy_parsed100percat_with_categories_june23.csv")
tiny_trigger = pd.read_csv("combined_chatgpt_words.csv", header=None).squeeze().tolist()
```

```python
tiny_revs.head()
```

```python
import re

# Define p_t computation (trigger word frequency)
def compute_pt(text, trigger_set):
    tokens = re.findall(r'\w+', str(text).lower())
    if not tokens:
        return 0.0
    return sum(1 for t in tokens if t in trigger_set) / len(tokens)

# Step 1: Filter baseline window
prechatgpt_revs = tiny_revs[tiny_revs["timestamp"] < "2022-11-01"]
#

```

```python
# Step 2: Compute p_t
prechatgpt_revs["p_t"] = prechatgpt_revs["plain_text"].apply(lambda txt: compute_pt(txt, tiny_trigger))
```

```python
prechatgpt_revs
```

```python

# Step 3: Compute baseline q
q = prechatgpt_revs["p_t"].mean()
print(f"Baseline q: {q:.6f}")
```

```python
# Step 4: Compute delta for all revisions
tiny_revs["p_t"] = tiny_revs["plain_text"].apply(lambda txt: compute_pt(txt, tiny_trigger))
tiny_revs["lexical_spike_delta"] = tiny_revs["p_t"] - q
```

```python
tiny_revs.to_csv("after_spacy_parsed100percat_with_categories_june23_with_delta.csv", index=False)
tiny_revs.to_pickle("after_spacy_parsed100percat_with_categories_june23_with_delta.pkl")
```
